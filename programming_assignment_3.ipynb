```python
#Q1 Write a Python Program to Check if a Number is Positive, Negative or Zero?
```

```python
def check(x):
    if(x>0):
        print("{} is postive".format(x))
    elif(x==0):
        print("{} is zero".format(x))
    else:
        print("{} is negative".format(x))
        
```

```python
check(0)
```

```python
check(-2)
```

```python
check(2.34)
```

```python
#Q2 Write a Python Program to Check if a Number is Odd or Even?
```

```python
def even_or_odd(x):
    if(x%2==0):
        return "{} is even".format(x)
    return "{} is odd".format(x)
```

```python
even_or_odd(3)
```

```python
even_or_odd(4)
```

```python
#Q3 Write a Python Program to Check Leap Year?
```

```python
def CheckLeap(Year):
    if((Year % 400 == 0) or(Year % 100 != 0) and (Year % 4 == 0)):
        print("Given Year is a leap Year")
    else:
        print ("Given Year is not a leap Year")  
```

```python
CheckLeap(2000)
```

```python
CheckLeap(1900)
```

```python
#Q4.Write a Python Program to Check Prime Number?
```

```python
def is_prime(x):
    a=[x%i for i in range(1,x+1) if x%i==0]
    b=len(a)
    if(b==2):
        return True
    return False
```

```python
is_prime(7)
```

```python
is_prime(109)
```

```python
is_prime(108)
```

```python
#Q5.Write a Python Program to Print all Prime Numbers in an Interval of 1-10000?
```

```python
def is_prime_interval(x,y):
    c=[i for i in range(x,y+1) if is_prime(i)==True] # using the is_prime function above
    return c
 
```
